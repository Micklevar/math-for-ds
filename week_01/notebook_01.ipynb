## 2. Synthesis of Concepts

#### 1.- Vectors

Vectors are used in multiple fields such as physics, computer science, and mathematics. In physics, vectors are represented as arrows with direction and magnitude that can move freely in a plane. In computer science, they are often represented as lists of data. Generally, vectors are any enelemt beloging to a vector space. More specifically, vectors are special elements with two important properties:

- **Additivity:** The sum of two vectors $\vec{a}$ and $\vec{b}$ results in another vector $\vec{c}$ - ($\vec{a}$+$\vec{b}$ = $\vec{c}$)
- **Scaling:** Multiplying a vector by scalar results in another vector which is the same but scaled ($\lambda$ $\vec{a}$ = $\vec{b}$)

The combination of these two properties is known as a linear combination, and these are the most important for considering a vector element, although in addition to these there are eight axioms wich determine what we can consider as a vector

#### 2.- Matrix

A matrix is essentially the result of a set of vectors. More technically, a matrix is a two-dimensional array of numbers, symbols, or expressions arranged in "m" rows and "n" columns. For example, a 3x3 matrix has 3 rows and 3 colimns. These types of mtrices are known as square matrices. There are other types of matrices, column matrices (vectors), diagonal matrices, upper triangular matrices, lower triangular matrices, and the zero matrix

#### 3.- Linear Transformation

To discuss linear transformation, it first necessary to talk about the elements that are part of linear transformation. Some of these have already been mentined, such as vectors and matrices; however, they will be mentioned in summary form along with the rest of the elements and concepts:

- **Vectors:** Special elements that are part of a vector space and have the properties of additivity and scaling.
- **Matrix:** A two-dimentional set of numbers, symbols or expressions arranged in m rows and n columns.
- **Basis:** A generating set of vector space.
- **Linear combination:** The set of operations of matrix addition and scalar multiplication.

With this in mind, we can then define a linear transformation as a transformation of the space itself, the space in wich the bases are transformed from the matrices. For a transformation to be considered linear, it must maintain two important properties:

1. It must maintain its origin without moving any other location.
2. The lines must remain parallel and evenly spaced.

### Product point manual

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Product point manual
a = np.array([1,5])
print(f"Vector a = {a}")
b = np.array([2,1])
print(f"Vector b = {b}")

print(a[0], "x", b[0], "+", a[1], "x", b[1])
c = a[0] * b[0] + a[1] * b[1] 
    
print(f"Dot product between the vectors a and b = {c}")
```

```python
# Function to generate graphs

def draw_vectors(vectors, colors=['green', 'red', 'blue']):
    plt.figure(figsize=(6,6))
    plt.axhline(0, color='black', lw=1) 
    plt.axvline(0, color='black', lw=1) 
    plt.grid(True, linestyle='--', alpha=0.6)

    for i, v in enumerate(vectors):
        plt.quiver(0, 0, v[0], v[1], color=colors[i % len(colors)], 
                   angles='xy', scale_units='xy', scale=1, label=f'Vector {i+1}')

    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.legend()
    plt.show()
```

```python
vectors_to_plot = [a,b]

draw_vectors(vectors_to_plot)
```

### Euclidean norm of a vector

```python

def euclidean_norm(vector):

    v = np.array(vector) # transform list to numpy array

    squares = v**2 

    sum_squares = np.sum(squares)

    norm = np.sqrt(sum_squares)

    return norm
```

```python
a = [1,5]

print(euclidean_norm(a))
```

### Matrix times a vector

```python
# Multiplication of a matrix by a vector

a = np.array([1,0])
b = np.array([[2,1], # matrix 2x2
            [3,2]])


c_x = a[0] * b[0,0] + a[1] * b[0,1]
c_y = a[0] * b[1,0] + a[1] * b[1,1]

c = np.array([c_x, c_y])

print(f"Original vector a = {a}")
print(f"Transformed vector c = {c}")
```

```python
vectors_to_plot = [a,c]
draw_vectors(vectors_to_plot)
```

### Multiplication of a matrix by vector

```python
rng = np.random.default_rng(seed=42)


list_vectors = []

while len(list_vectors)< 10 :
    enteros = np.array(rng.integers(0, 10, size=2))
    list_vectors.append(enteros)

print(list_vectors)
```

```python
theta = np.radians(45)
rotation = np.array([[np.cos(theta), -np.sin(theta)],
                    [np.sin(theta), np.cos(theta)]])
```

```python
transformed_vectors = []
for i in list_vectors:
    c = np.array(rotation@i)

    transformed_vectors.append(c)
    print(c)
```

```python
def draw_vectors(vectors, colors=['blue', 'orange']):
    plt.figure(figsize=(8, 8))
    plt.axhline(0, color='black', lw=1) 
    plt.axvline(0, color='black', lw=1) 
    plt.grid(True, linestyle='--', alpha=0.6)

    mitad = len(vectors) // 2

    for i, v in enumerate(vectors):
        current_color = 'blue' if i < mitad else 'orange'
        
        if i == 0:
            current_label = 'Originals'
        elif i == mitad:
            current_label = 'Transformed'
        else:
            current_label = None

        plt.quiver(0, 0, float(v[0]), float(v[1]), 
                   color=current_color, 
                   angles='xy', scale_units='xy', scale=1, 
                   label=current_label)

    all_coords = np.array(vectors)
    limit = np.max(np.abs(all_coords)) + 1
    
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.xlim(-limit, limit)
    plt.ylim(-limit, limit)
    plt.title("Vector comparison chart after applying a 45-degree transformation")
    plt.legend() 
    plt.show()
```

```python
colors = ['blue'] * 10 + ['orange'] * 10

draw_vectors(list_vectors + transformed_vectors, colors=colors)
```
