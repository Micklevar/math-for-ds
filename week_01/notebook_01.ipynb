## Fundamental Concepts

#### 1.- Vectors

Vectors are used in multiple fields such as physics, computer science, and mathematics. In physics, vectors are represented as arrows with direction and magnitude that can move freely in a plane. In computer science, they are often represented as lists of data. Generally, vectors are any enelemt beloging to a vector space. More specifically, vectors are special elements with two important properties:

- **Additivity:** The sum of two vectors $\vec{a}$ and $\vec{b}$ results in another vector $\vec{c}$ - ($\vec{a}$+$\vec{b}$ = $\vec{c}$)
- **Scaling:** Multiplying a vector by scalar results in another vector which is the same but scaled ($\lambda$ $\vec{a}$ = $\vec{b}$)

The combination of these two properties is known as a linear combination, and these are the most important for considering a vector element, although in addition to these there are eight axioms wich determine what we can consider as a vector

#### 2.- Matrix

A matrix is essentially the result of a set of vectors. More technically, a matrix is a two-dimensional array of numbers, symbols, or expressions arranged in "m" rows and "n" columns. For example, a 3x3 matrix has 3 rows and 3 colimns. These types of mtrices are known as square matrices. There are other types of matrices, column matrices (vectors), diagonal matrices, upper triangular matrices, lower triangular matrices, and the zero matrix

#### 3.- Linear Transformation

To discuss linear transformation, it first necessary to talk about the elements that are part of linear transformation. Some of these have already been mentined, such as vectors and matrices; however, they will be mentioned in summary form along with the rest of the elements and concepts:

- **Vectors:** Special elements that are part of a vector space and have the properties of additivity and scaling.
- **Matrix:** A two-dimentional set of numbers, symbols or expressions arranged in m rows and n columns.
- **Basis:** A generating set of vector space.
- **Linear combination:** The set of operations of matrix addition and scalar multiplication.

With this in mind, we can then define a linear transformation as a transformation of the space itself, the space in wich the bases are transformed from the matrices. For a transformation to be considered linear, it must maintain two important properties:

1. It must maintain its origin without moving any other location.
2. It must maintain its axes equidistants.
