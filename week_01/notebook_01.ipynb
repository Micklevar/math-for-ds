## Fundamental Concepts

#### 1.- Vectores

Un vector puede ser cualquier cosa que siga dos reglas simples, puede sumarse entre ellos es decir un vector A y un vector B si se suman pueden dar un vector C (A+B=C) y ademas pueden estos multiplicarse por un escalar (C*A=D). Esta definicion es un poco abierta y deja que muchas cosas puedan ser tomadas como vectores es esto exactamente lo que produce que muchas veces se hable de vectores en distintas ramas, como la fisica donde por lo general se habla de vectores como flechas que tienen una longitud y una direccion, mientras que en el lado de la matematica se trata mas a los vectores como posiciones en un espacio de n dimensiones igual representado con flechas pero que esta vez tienen una longitud. En ambos casos siguen esta regla de la aditividad y la escalacion, es decir la suma entre vectores y la multiplicacion por escalar. En general entonces puede ser llamado un vector a cualquiera que siga los conocidos hasta ahora 8 axiomas, hay muuuchos conceptos matematicos que pueden ser tomados para resolverlos o ser vistos como vectores como son las funciones, a todos estas maneras de ver los vectores se les conoce como Espacio vectorial, lo que puede ser confundido con "espacio generado" pero este es mas bien el espacio o cuadricula que genera todos los posibles puntos que puede tomar un vector despues de una transformacion

#### 2.- Matrices

Cuando hablamos de matrices estamos tambien hablando de vectores, ya que las matrices son en si conjuntos de vectores ordenados, estos tienen las caracteristicas de las filas y las columnas, siendo las columnas las que nos dicen en cuantas dimensiones estaremos situados, una matriz cuadrada de 2x2 estaria situada en el plano x y que todos conocemos, y una de 3x3 estaria situando un vector en un plano tridimensional con x,y,z. Hay muchos tipos de matrices como la matriz identidad, la nula, matriz diagonal, triangular superior, triangular inferior entre otras, cada una tiene una caracteristica importante, aunque el libro aun no explora eso en el capitulo 1 y en la serie de videos se vio todo de una perspectiva mas geometrica no tanto de calculo. Sin embargo otra de las caracteristicas de las matrices es que sirven para resolver ecuaciones, esto tambien entra en lo que hablabamos del espacio generado, teniendo un sistema de ecuaciones de 3x3 podriamos representarlo como una matriz de 3 ecuaciones donde se encuentran sus escalares, es decir cualquier numero, luego una matriz columna o un vector con las tres incognitas x,y,z y por ultimo una matriz vector o columna con los resultados podriamos resolver las ecuaciones usando metodos de algebra lineal aplicados a matrices.

#### 3.- Transformacion lineal

Una transformacion lineal es la deformacion del espacio n dimensional que se produce al multiplicar a nuestras bases por una matriz, dos conceptos que hay que tener en cuenta es que en estas transformaciones las caracteristicas principales seran que las distancias se mantendran siempre equidistantes y se mantendra siempre el vector 0,0 osea el centro, por lo demas todo puede modificarse. puede voltearse y presentar numeros negativos o incluso podria volverse una sola linea en donde todo el espacio se comprimiria.
